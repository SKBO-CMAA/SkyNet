# Genera el listado de la carpeta Estaciones
Index2.html

```python
import os
import json

carpeta_base = "Estaciones"

def construir_arbol(ruta):
    nodo = {
        "nombre": os.path.basename(ruta),
        "tipo": "carpeta",
        "contenido": []
    }

    for elemento in sorted(os.listdir(ruta)):
        ruta_completa = os.path.join(ruta, elemento)

        if os.path.isdir(ruta_completa):
            nodo["contenido"].append(construir_arbol(ruta_completa))
        else:
            nodo["contenido"].append({
                "nombre": elemento,
                "tipo": "archivo",
                "ruta": ruta_completa.replace("\\","/")
            })

    return nodo


estructura = construir_arbol(carpeta_base)

with open("estructura_estaciones.json","w",encoding="utf8") as f:
    json.dump(estructura,f,indent=2,ensure_ascii=False)

print("Árbol de carpetas generado")
```

# Index.html

```python
import os
import json

carpeta = "Estaciones"

def construir_arbol(ruta):
    
    items = []
    
    for nombre in sorted(os.listdir(ruta)):
        
        path = os.path.join(ruta, nombre)
        
        if os.path.isdir(path):
            
            items.append({
                "tipo": "carpeta",
                "nombre": nombre,
                "contenido": construir_arbol(path)
            })
        
        else:
            
            items.append({
                "tipo": "archivo",
                "nombre": nombre,
                "ruta": path.replace("\\","/")
            })
    
    return items


estructura = construir_arbol(carpeta)

with open("estaciones.json","w",encoding="utf8") as f:
    json.dump(estructura,f,indent=2,ensure_ascii=False)

print("estructura completa generada")
```
