```python
import fitz
import os
from PIL import Image

pdf_dir="pdf"
img_dir="img"

for pdf in os.listdir(pdf_dir):

    if pdf.endswith(".pdf"):

        doc=fitz.open(f"{pdf_dir}/{pdf}")

        page=doc.load_page(0)

        pix=page.get_pixmap(matrix=fitz.Matrix(2,2))

        png_name=pdf.replace(".pdf",".png")
        webp_name=pdf.replace(".pdf",".webp")

        png_path=f"{img_dir}/{png_name}"
        webp_path=f"{img_dir}/{webp_name}"

        pix.save(png_path)

        img=Image.open(png_path)
        img.save(webp_path,"WEBP",quality=80)

        os.remove(png_path)

        print("Generado:",webp_name)
```
